## Visualizations


```python
corr = df.corr()

mask = np.triu(np.ones_like(corr, dtype=bool))

sns.heatmap(df.corr(numeric_only = True),annot= True,mask=mask)
plt.figure(figsize= (20,6))
plt.show()
```

**Analysis:**

* positive correlation (0.88) between quatlity of sleep and sleep duration
* negative correlation (-0.81) between stress level and sleep duration
* negative correlation (-0.9) between stress level and quality of sleep
* slight negative correlation (-0.66) between heart rate and quality of sleep
* slight positive correlation (0.67) between heart rate and stress level
* positive correlation (0.77) between daily steps and physical activity level




```python
cols_to_plot = ['Sleep Duration', 'Stress Level', 'Physical Activity Level', 'Quality of Sleep']
sns.pairplot(df[cols_to_plot],kind='reg',plot_kws={'line_kws':{'color':'red'}})
```

```python
# create list of variables to plot
list_of_col = ['Gender', 'Occupation', 'Sleep Disorder', 'BMI Category']

# creating histograms for each categorcial column in list_of_col variable

for col in list_of_col:
  plt.hist(df[col])
  plt.title(col)
  plt.xticks(rotation=75)
  plt.xlabel('categories')
  plt.ylabel('values')
  plt.show()
```

**Question:**

Do high levels of physical activity level impact sleep?

```python
# high levels of physical activity = 60+ mins

# filtering original df by physical activity level
high_physical_activity = df.loc[df['Physical Activity Level'] >= 60]
high_physical_activity
```

```python
sns.lmplot(x='Physical Activity Level', y='Quality of Sleep', data=high_physical_activity)
plt.title('Physical Activity Level and Quality of Sleep')
```

***Analysis:***

* Unable to find a strong correlation between high physical activity levels and quality of sleep




**Question:**

How does stress level impact quality of sleep?

```python
# creating scatter plot for relationship between stress level and quality of sleep
sns.lmplot(x='Stress Level', y='Quality of Sleep', data=df)
plt.title('Stress Level and Quality of Sleep')
```

***Analysis:***

* As participants stress level increased, their quality of sleep decreased




**Question:**

How did occupation effect quality of sleep?

```python
# creating a new df based on the occupations of participants and taking the mean of their quality of sleeep rating
occupation_df = df.groupby('Occupation')['Quality of Sleep'].mean()
occupation_df
```

```python
# plotting relationship between occupation and quality of sleep
occupation_df.plot(x= ['Occupation'], y= ['Quality of Sleep'])
plt.title('Occupation and Quality of Sleep')
plt.xticks(rotation=45)
plt.xlabel('Occupation')
plt.ylabel('Quality of Sleep')
plt.show()
```

***Analysis:***

* Engineers reported the highest quality of sleep on average at 8.4
* Sales representatives reported the lowest quality of sleep rating on average at 4
